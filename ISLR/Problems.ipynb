### Notes 

The least squares approach chooses the `beta_0` and `beta_1` to minimize the RSS.  They have the following formulas:

```
def beta_1(target, predictor):
    numerator = []
    denominator = []
    for x, y in zip(predictor, target):
        num_val = (x - np.mean(predictor)) * (y - np.mean(target))
        denom_val = (x - np.mean(predictor))**2
        numerator.append(num_val)
        denominator.append(denom_val)
    return np.sum(numerator) / np.sum(denominator)

def beta_0(target, predictor, beta_1):
    return np.mean(target) - beta_1 * np.mean(predictor)
```

Lets see this in action with the advertising dataset 

````python
df = pd.read_csv('data/Advertising.csv', index_col = 0)
df.head()
````

I am going to use TV to predict Sales 

````python
def beta_1(target, predictor):
    numerator = []
    denominator = []
    for x, y in zip(predictor, target):
        num_val = (x - np.mean(predictor)) * (y - np.mean(target))
        denom_val = (x - np.mean(predictor))**2
        numerator.append(num_val)
        denominator.append(denom_val)
    return np.sum(numerator) / np.sum(denominator)
````

````python
def beta_0(target, predictor, beta_1):
    return np.mean(target) - beta_1 * np.mean(predictor)
````

````python
target = df['sales']
predictor = df['TV']
````

````python
beta_1_coef = beta_1(target, predictor)
beta_0_coef = beta_0(target, predictor, beta_1_coef)
beta_1_coef, beta_0_coef
````

As a sanity check I am going to compare it vs statsmodels 

````python
import statsmodels.api as sm

predictor = sm.add_constant(predictor)
model = sm.OLS(target, predictor).fit()
model.summary()
````

````python
# load in auto dataset 
df = pd.read_csv('data/Auto.csv')
df.head()
````

````python
# horsepower is an object
df.dtypes
````

````python
df = df[df['horsepower'] != '?']
df['horsepower'] = df['horsepower'].astype(int)
````

````python
from statsmodels.formula.api import ols
formula = 'mpg ~ horsepower'
model = ols(formula=formula, data=df).fit()
model.summary()
````

<b>i. Is there a relationship between the predictor and the response?</b>

There is a relationship between the predictor and the response.  I can tell this by the high R^2 and the F-statistic being large. 

<b> ii. How strong is the relationship between the predictor and
the response?</b>

Horsepower accounts for 61% of the variation in mpg

<b> iii. Is the relationship between the predictor and the response
positive or negative? </b>

The relationship is negative.  I can see this due to the coefficient for horsepower being -0.1578.  This means as the horsepower increases, the mpg decreases

<b> iv. What is the predicted mpg associated with a horsepower of
98? What are the associated 95 % confidence and prediction
intervals? </b>

````python
model.predict({'horsepower': 98})
````

````python
model.conf_int()
````

````python
plt.scatter(df['horsepower'], df['mpg'])
plt.plot([df['horsepower'].min(), df['horsepower'].max()], 
        model.predict({'horsepower': [df['horsepower'].min(), df['horsepower'].max()]}),
        lw = 5, color = 'k')
plt.xlabel('horsepower')
plt.ylabel('mpg')
````
