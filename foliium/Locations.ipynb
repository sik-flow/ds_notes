```python
import time
```

```python
lat = []
long = []
name = []
errors = []
locator = Nominatim(user_agent="myGeocoder")
for counter, x in enumerate(df['location']):
    try: 
        if counter % 10 == 0:
            print(f'Completed {counter} our of {len(df["location"])}')
        location = locator.geocode(x)
        lat.append(location.latitude)
        long.append(location.longitude)
        name.append(x)
        time.sleep(1)
    except:
        errors.append(x)
```

```python

import numpy as np

size = 100
lons = np.random.randint(-180, 180, size=size)
lats = np.random.randint(-90, 90, size=size)

locations = list(zip(lat, long))
popups = ['lon:{}<br>lat:{}'.format(lon, lat) for (lat, lon) in locations]
```

```python
icon_create_function = """\
function(cluster) {
    return L.divIcon({
    html: '<b>' + cluster.getChildCount() + '</b>',
    className: 'marker-cluster marker-cluster-large',
    iconSize: new L.Point(20, 20)
    });
}"""
```

```python
from folium.plugins import MarkerCluster


m = folium.Map(
    location=[np.mean(lats), np.mean(lons)],
    tiles='CartoDB dark_matter',
    zoom_start=2
)

marker_cluster = MarkerCluster(
    locations=locations, popups=name,
    name='1000 clustered icons',
    overlay=True,
    control=True,
    #icon_create_function=icon_create_function
)

marker_cluster.add_to(m)

folium.LayerControl().add_to(m)
m
```
